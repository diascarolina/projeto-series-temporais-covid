# Dados Sobre Casos e Óbitos da COVID-19

```python
df = pd.read_csv('../data/caso_full.csv.gz')
```

```python
dados_ce = df.query('state == "CE"')
```

```python
dados_ce = dados_ce.query('place_type == "state"').drop('city', axis = 1)
```

```python
colunas = ['date', 'new_confirmed', 'new_deaths', 'last_available_confirmed', 'last_available_deaths',
           'epidemiological_week', 'estimated_population_2019']
```

```python
dados_ce = dados_ce[colunas]
```

```python
colunas_pt = ['data', 'novos_casos', 'novos_obitos', 'total_casos', 'total_mortes',
             'semana_epidemiologica', 'pop_estimada_2019']
```

```python
dados_ce.columns = colunas_pt
```
